```
import os
import sys

import numpy as np

dir_path = os.path.dirname(os.getcwd())
sys.path.append(os.path.join(dir_path, 'src'))

from pyhocon import ConfigTree, ConfigFactory
from bin import BinMapper
from gradient_did_tree import GradientDebiasedCausalTree
from boosting import Boosting
from dataset import Dataset
```

```
# load data
conf_path = os.path.join(dir_path, 'config/gbct.conf')
conf = ConfigFactory.parse_file(conf_path)
conf.put('$.dataset.data.path', [f'{dir_path}/data/binary_var_0.5/X.csv', f'{dir_path}/data/binary_var_0.5/Y.csv'])
conf.put('$.iterators', 50)
data = Dataset.new_instance(conf)
bin_mapper = BinMapper(conf)
# read the index
idx = np.load(os.path.join(os.path.dirname(conf.dataset.data.path[0]), 'tr_te.npy'), allow_pickle=True)

# just run fold 0
fold = 0
tr_data = data.sub_dataset(idx[fold, 0])
valid_data = data.sub_dataset(idx[fold, 1])
```

```
cf = Boosting(GradientDebiasedCausalTree, conf, bin_mapper)
cf.fit(data, valid_data)
```
